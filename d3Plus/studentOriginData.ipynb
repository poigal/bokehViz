```python
df = pd.read_csv("20171129125141018050EDA0243753659155.csv",header=-1)
df.head()
```

```python
# remove first column
df = df.iloc[:,1:]
df.head()
```

```python
# Assigncolumn names -Years
r = [str(i) for i in list(range(2000,2017))]
n=['countryOrigin']
n.extend(r)
df.columns = np.array(n,dtype=object)
df.columns
```

```python
#df = df.iloc[:32,:]
df.head()
```

```python
# Conver data to long-form
df = df.melt(id_vars="countryOrigin",var_name="Year",value_name="StudCount")
df.head()
```

```python
# removeing EU 27 (redundant) and Ireland
df = df[~df['countryOrigin'].isin(['EU 27','Ireland'])]
```

```python
df.tail()
```

```python
# removing string '(1)' from the country names
import re
df.countryOrigin = [re.sub(r'[ (1)]','',x) for x in df.countryOrigin]
df.tail()
```

```python
df.to_csv("studentOrigin.csv",index=False)
```
